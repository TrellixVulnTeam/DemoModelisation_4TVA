```
import pandas as pd

f_eveler = "/content/drive/My Drive/EEW/Data/eveler/narbonne_mars2020.pkl"
f_sscomp = "/content/drive/My Drive/EEW/Data/resu_sous_comptage/Conso_Estimee_Narbonne_Mars2020_2.pkl"




df_eveler = pd.read_pickle(f_eveler)
df_sscomp = pd.read_pickle(f_sscomp) 

# Consommation journalière en kWh/jour Eveler

df_eveler_conso_jour = df_eveler['P_ACT_KW'].resample('D').mean()*24
# somme des consommations du sous-comptage

df_sscomp_conso_jour = df_sscomp.sum(axis=1)


import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(
    go.Scatter(x=df_sscomp_conso_jour.index, y=df_sscomp_conso_jour.values,name='Somme sous-compt. virtuels')

)

fig.add_trace(
    go.Scatter(x=df_eveler_conso_jour.index, y=df_eveler_conso_jour.values,name='Conso Eveler journalières')
)

fig.update_layout(height=600, width=800, title_text="Conso Eveler/Total sous-comptage virtuel")
fig.show()
```
